# Homo and Heteroscedasticity exploration

Here I try to introduce the concepts of homo and heteroscedasticity.

```python
# general libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# functions

def plot(x,y,title, x_label = 'x', y_label = 'y', **kwargs):
    plt.plot(x,y,'.',**kwargs)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

### Homoscedasticity

This is the assumption that the errors in a model (or **noise**) do not vary with the independent (predictor) variables.

For example, consider this:

```python
x = np.linspace(0,2*np.pi,100)
y_base = np.sin(x)
y = y_base + np.random.normal(0,0.3,100)

plot(x,y,'a plot of a noisy sine wave with $\mu=0$, $\sigma = 0.3$')

error = np.abs(y - y_base)
plot(x,error, 'a plot of the errors against the dependent variable', y_label = 'error')
```

From the above, we can see that the data is generally 'uniform'. I.e. an increase in the predictor variable x is not really affecting the size of the errors.

### Heteroscedasticity

This is when the **noise** in the data is no longer the same across the predictor variables, but rather increases or decreases with the predictor.

For example, consider this:

```python
x = np.linspace(0,2*np.pi,100)
y_base = np.sin(x)

y = []
for i,y_val in enumerate(y_base):
    y.append(y_base[i] + np.random.normal(0,i/100,1))

plot(x,y,'a plot of a noisy sine wave with $\mu=0$, $\sigma = increasing$')

error = np.abs(y - y_base.reshape(-1,1))
plot(x,error, 'a plot of the errors against the dependent variable', y_label = 'error')
```

Above, we see that the errors are increasing with the predictor variable. This is problematic for any model that assumes that the erros are normally distributed.

In fact, let's try to sketch the histogram for this to see what it looks like

```python
plt.hist(error)

# see if a transformation can help?

plt.hist(np.power(error,1/4),alpha = 0.8)
plt.title('histogram for heteroscedastic errors')
plt.xlabel('$\ln$ error')
plt.ylabel('frequency')

plt.legend(['without transformation','with a transformation'])
```
