# SVR - Support Vector Regression

This notebook will cover the concept of support vector regresssion. It assumes a knowledge of support vector machines.

# Libraries

```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt


from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

# Why Support Vector Regression?

Before we begin discussing how SVR works, let's first understand why there is a need for this in the first place.

Consider a linear regression problem, as shown by the data below:

```python
x = np.linspace(0, 100, 100)
y = x + np.random.normal(0,10,100)
y[50] *= 3
y[90:100] *= 3
print(y.shape)
print(x.shape)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y,'g.')
fig.legend(['data points'])
```

```python
models = []
yhats = []

# reshape x and y
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
models.append(LinearRegression())
models[0].fit(x,y)
yhats.append(models[0].predict(x))
```

```python
ax.plot(x,yhats[0],'.')
fig.legend(['data points','linear regression'])
fig
```

As you can see, the line is heavily affected by the presence of outliers. This is because of the way the linear regression is calculated: it tries to minimise the mean squared error, which is highly impacted by outliers.

Let's see what SVR can do instead.

```python
models.append(SVR(kernel = 'linear'))
models[1].fit(x,y)
yhats.append(models[1].predict(x))
```

```python
ax.plot(x,yhats[1],'.')
fig.legend(['data points','linear regression','SVR'])
fig
```

We can see that the SVR has almost an excellent fit. This is because instead of trying to fit the points such that it minimises error, it tries to fit them such that it covers the highest number of points. It's quite a smart trick.

There is the problem that SVM can be quite prone to overfitting however. That said, it performs fairly well in high dimensional datasets.
